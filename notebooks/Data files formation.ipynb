```python
dataset = pd.read_csv('../data/external/DrDoS_DNS.csv', names = cols, skiprows = 1).drop(['1'], axis=1)
dataset
```

```python
# split data by the Label column value
benign_temp = dataset[dataset['Label'] == 'BENIGN']
benign_temp = benign_temp.reset_index(drop=True)
ddos_temp = dataset[dataset['Label'] == 'DrDoS_DNS']
ddos_temp = ddos_temp.reset_index(drop=True)
```

```python
# split by the size

temp_benign_train = benign_temp[:2000]
temp_benign_test = benign_temp[2000:3000]
temp_benign_val = benign_temp[3000:]
```

```python
print(len(temp_benign_val), len(temp_benign_test), len(temp_benign_train))
```

```python
# split by size

temp_ddos_train = ddos_temp[:4000000]
temp_ddos_test = ddos_temp[4000000:5000000]
temp_ddos_val = ddos_temp[5000000:]
```

```python
print(len(temp_ddos_val), len(temp_ddos_test), len(temp_ddos_train))
```

```python
# shuffling

from sklearn.utils import shuffle

training = shuffle(pd.concat([temp_benign_train, temp_ddos_train], ignore_index = True))
training.reset_index(inplace=True, drop=True)
#training = training.sample(frac=1).reset_index(drop=True)
testing = shuffle(pd.concat([temp_benign_test, temp_ddos_test], ignore_index = True))
testing.reset_index(inplace=True, drop=True)
#testing = testing.sample(frac=1).reset_index(drop=True)
validation = shuffle(pd.concat([temp_benign_val, temp_ddos_val], ignore_index = True))
validation.reset_index(inplace=True, drop=True)
#validation = validation.sample(frac=1).reset_index(drop=True)
```

```python
testing['Label'] = testing['Label'].replace({ 'DrDoS_DNS': 1})
testing['Label'] = testing['Label'].replace({ 'BENIGN': 0})
```

```python
testing_combined = pd.DataFrame()
testing_combined['combined'] = testing[cols2].apply(lambda row: ','.join(row.values.astype(str)), axis=1)
```

```python
# creating and saving CSV files

training.to_csv('../data/interim/training.csv', index = False, columns = cols2)
testing.to_csv('../data/interim/testing.csv', index = False, columns = cols2)
testing_combined.to_csv('../data/interim/testing_prod.csv', index = False)
validation.to_csv('../data/interim/validation.csv', index = False, columns = cols2)
```
